```python
for col in cat_data:
    clc = df[col].value_counts()
    plt.figure(figsize=(8, 6))
    plt.bar(clc.index, clc.values) 

    plt.title(f"{col} Distribution") 
    plt.xlabel(f'{col}') 
    plt.ylabel('Counts') 

    print(clc)
    plt.show()  
    print("\n\n")
```
